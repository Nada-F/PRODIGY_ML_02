```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

```

```python
Data = pd.read_csv('/content/drive/MyDrive/Mall_Customers.csv')
Data
```

```python
# Inspect the dataset
print(Data.head())
```

```python
Data = Data.drop(['CustomerID', 'Gender',], axis=1)
Data
```

```python
# Plot scatter plot for 'Annual Income (k$)' and 'Spending Score (1-100)'
plt.figure(figsize=(10, 6))
plt.scatter(Data['Annual Income (k$)'], Data['Spending Score (1-100)']) # Changed df to Data
plt.title('Annual Income vs Spending Score')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.show()
```

```python
scaler = StandardScaler()
Data_scaled = scaler.fit_transform(Data)
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(Data_scaled)
    wcss.append(kmeans.inertia_)
```

```python
# Plot the elbow method results
plt.figure(figsize=(10, 6))
plt.plot(range(1, 11), wcss, marker='o')
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS (Within Cluster Sum of Squares)')
plt.show()
```

```python
kmeans = KMeans(n_clusters=5, init='k-means++', max_iter=300, n_init=10, random_state=42)
clusters = kmeans.fit_predict(Data_scaled)
```

```python
# Assuming you want to perform PCA with 2 components
import pandas as pd
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
df_pca = pca.fit_transform(Data_scaled)

pca_Data = pd.DataFrame(df_pca, columns=['PCA1', 'PCA2'])
```

```python
pca = PCA(n_components=2)
Data_pca = pca.fit_transform(Data_scaled)
```

```python
pca_Data['Cluster'] = clusters
```

```python
# Plot the clusters
plt.figure(figsize=(8, 4))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', palette='Set1', data=pca_Data, s=100) # Changed pca_df to pca_Data
plt.title('Customer Segmentation based on K-means Clustering')
plt.show()
```
